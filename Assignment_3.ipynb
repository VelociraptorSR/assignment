3.  What are the different ways to define and call a function in Python?



```python
#Standard function definition (using def):
def add(a, b):
    return a + b
add(5, 10)
```

```python
#Function with default parameters:
def greet(name="Guest"):
    return f"Hello, {name}!"

greet(name="Alice")
greet()  # uses default "Guest"

```

```python
#Variable-length arguments:
def print_numbers(*args):
    for num in args:
        print(num)

def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
print_numbers(1, 2, 3, 4)
print_details(name="Alice", age=25)
```

```python
#Lambda functions
add = lambda x, y: x + y

add(7, 8)
```

```python
#Functions inside a class (methods):
class Calculator:
    def multiply(self, a, b):
        return a * b
calc = Calculator()
calc.multiply(3, 4)
```

5.  What are iterators in Python and how do they differ from iterables?
  - Iterable ->
  - An iterable is any object you can loop over (like using for loops).
Examples: lists, tuples, strings, sets, dictionaries, etc.
  - Basically, if an object has a method called __iter__(), it's an iterable.
  - Iterators ->
  - An iterator is an object that actually produces the next value when you call next() on it. It remembers where it left off.
  - If an object has both __iter__() and __next__() methods, it’s an iterator.

6.  Explain the concept of generators in Python and how they are defined.
  - A generator is a special type of iterator.
  - Instead of building and returning all the values at once, it yields them one at a time as you need them.
  - It saves memory because it generates values on the fly.

7. What are the advantages of using generators over regular functions?
  - They save memory — they don't store the whole sequence in memory.
  - They are lazy — they calculate values only when needed.
  - Perfect for handling large datasets or infinite streams of data.

8. What is a lambda function in Python and when is it typically used?
  - A lambda function is a small, anonymous (nameless) function.
  - It’s defined using the keyword lambda instead of def.
  - It can have any number of inputs, but only one expression (the result of that expression is automatically returned).
  - When you need a quick, small function for short-term use.
  - When you don't want to formally define a full function with def.
  - Especially useful inside functions like map(), filter(), and sorted(), etc.

9. Explain the purpose and usage of the map() function in Python.
  - map() is a built-in function that applies a given function to every item in an iterable (like a list, tuple, etc.) and returns a new map object (which is an iterator).
  - You can map multiple iterables together too.
  - map() lets you apply a function to every item in an iterable quickly and cleanly.

10. What is the difference between map(), reduce(), and filter() functions in Python?

  - filter() -> Keeps only the items where the function returns True.

1.  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def even_sum(l):
  sum = 0
  for i in l:
    if i%2==0:
      sum = sum + i
  return sum

l = [1,2,3,4,5,6,7,8,9]
print(even_sum(l))
```

3.  Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number.

```python
def square(l):
  sq = []
  for i in l:
    sq.append(i**2)
  return sq

l = [2,4,3,5,7,8]
print(square(l))
```

```python
def prime(num):
  if 1<=num<=200:

    if num==1 or num==2:
      return 'yes number is prime'
    else:
      flag = True
      for i in range(2,num):
        if num%i==0:
          flag = False
      if flag == True:
        return 'yes number is prime'
      else:
        return 'number is not prime'

  else:
    print('enter the number between 1 to 200')

n = int(input('enter any number between 1 to 200'))
print(prime(n))
```

6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def ex(n):
  for i in range(1,n+1):
    yield 2**i

for i in ex(5):
  print(i)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line

for line in read_file('example.txt'):
    print(line)
```

8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

```python
l = [(1,3),(2,4),(1,2),(4,2,5)]
s = sorted(l,key=lambda x:x[1])
print(s)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
def celsius(c):
  return (c*9/5) + 32

temps = [0, 20, 37, 100]

fah = list(map(celsius,temps))
print(fah)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string

```python
s = 'saumya rajpoot'
vow = set('aeiouAEIOU')
without_vow = list(filter(lambda x:x not in vow,s))
print(''.join(without_vow))
```
