1. What is hypothesis testing in statistics
  - Hypothesis testing is a formal statistical method used to make decisions or inferences about a population based on sample data. To test a claim or assumption (called a hypothesis) about a population parameter (like mean, proportion, or variance).

2. What is the null hypothesis, and how does it differ from the alternative hypothesis
  - 1. Null Hypothesis (H₀):
  - It is the default or status quo assumption.
  - It states that there is no effect, no difference, or no change.
  - The goal of hypothesis testing is to test whether there is enough evidence to reject the null hypothesis
  - 2. Alternative Hypothesis (H₁ or Hₐ):
  - This is what you want to prove or provide evidence for.
  - It states that there is a difference, an effect, or a change.
  - You accept this if there's enough evidence to reject H₀.

3. What is the significance level in hypothesis testing, and why is it important
  - The significance level, denoted by α (alpha), is a threshold that helps you decide whether to reject the null hypothesis (H₀).

4. What does a P-value represent in hypothesis testing
  - The p-value is the probability of obtaining a test result as extreme as, or more extreme than, the one observed in your sample assuming that the null hypothesis is true.

5.  How do you interpret the P-value in hypothesis testing
  - The p-value tells you how likely your observed results (or more extreme ones) would be if the null hypothesis (H₀) were true.
  - Compare the p-value to your chosen significance level (α, usually 0.05)
  - If p-value ≤ α → The result is statistically significant
  - Reject H₀
  - Evidence supports the alternative hypothesis (H₁)
  - If p-value > α → The result is not statistically significant
  - Fail to reject H₀
  - Not enough evidence to support H₁

6. What are Type 1 and Type 2 errors in hypothesis testing
  - Type 1 - Rejecting the null hypothesis when it is actually true.
  - Type 2 - Failing to reject the null hypothesis when it is actually false.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing
  - 1. Two-Tailed Test
  - H₁ claims the parameter is not equal to a certain value.
  - Tests for an effect in both directions (greater or less than).
  - The rejection region is split between both tails of the distribution.
  -  2. One-Tailed Test
  - H₁ claims the parameter is greater than or less than a certain value.
  - Tests for an effect in one direction only.
  - The rejection region is only in one tail of the distribution.

8. What is the Z-test, and when is it used in hypothesis testing
  - A Z-test is a statistical test used to determine whether there is a significant difference between a sample statistic and a population parameter (or between two sample means), assuming the population follows a normal distribution.
  - Population standard deviation (σ)	- Known
  - Sample size (n)	- Usually large (n ≥ 30) or population is normal
  - Data type -	Normally distributed or approximately normal

9. How do you calculate the Z-score, and what does it represent in hypothesis testing
  - A Z-score (also called a standard score) tells you how many standard deviations a data point or sample statistic is from the population mean.
  - Z = sample mean -  population mean/(population standard deviation/sqrt of sample size)
  - It shows how extreme your sample mean is if the null hypothesis (H₀) were true.
  - A higher absolute Z-score means your result is less likely due to chance.
  - The Z-score is used to calculate the p-value, which helps decide whether to reject H₀.
  


10. What is the T-distribution, and when should it be used instead of the normal distribution
  - The T-distribution (also called Student's t-distribution) is a probability distribution that is similar to the normal distribution, but with heavier tails. This means it accounts for more variability, especially in small samples.
  -  The population standard deviation (σ) is unknown
  -  The sample size (n) is small (typically n < 30)
  -  The data is approximately normally distributed

11. What is the difference between a Z-test and a T-test
  - 1. Population Standard Deviation:
  - In a Z-test, the population standard deviation (σ) is known.
  - In a T-test, the population standard deviation is unknown and the sample standard deviation (s) is used instead.
  - 2. Sample Size:
  - The Z-test is typically used when the sample size is large (n ≥ 30).
  - The T-test is preferred when the sample size is small (n < 30).
  - 3. Distribution Used:
  - The Z-test relies on the standard normal distribution (Z-distribution).
  - The T-test uses the T-distribution, which has thicker tails to account for additional uncertainty.
  - 4. Tail Thickness:
  - The T-distribution has heavier tails than the Z-distribution, which means it is more conservative for small samples.
  - As the sample size increases, the T-distribution approaches the normal distribution, making both tests give similar results.

12. What is the T-test, and how is it used in hypothesis testing
  - A T-test is a statistical test used to determine whether there is a significant difference between:
  - A sample mean and a known or hypothesized population mean
  - The means of two independent groups
  - The means before and after a treatment in the same group
  - The T-test helps you test claims about population means.

13. What is the relationship between Z-test and T-test in hypothesis testing
  - Z-test is used when you know more (like population SD) or have a large sample.
  - T-test is used when you know less (only sample SD) or have a small sample.
  - Both are tools to do the same job: evaluate whether a sample result is statistically significant.

14. What is a confidence interval, and how is it used to interpret statistical results
  - A confidence interval (CI) is a range of values derived from sample data that is likely to contain the true population parameter (like the mean or proportion) with a certain level of confidence.
  - Estimates the true value of the population parameter.
  - Shows precision — narrower interval = more precise estimate.
  - Helps test significance — if the interval does not include the null value, the result is significant.
  - Related to p-values — a 95% CI that excludes 0 means p < 0.05.
  - Not a probability — 95% confidence means 95 out of 100 such intervals will capture the true value if we repeat the study.

15. What is the margin of error, and how does it affect the confidence interval
  - The margin of error (MOE) is the maximum expected difference between the sample statistic and the true population parameter. It shows how accurate or uncertain the estimate is.
  - Determines the width of the confidence interval:
  - CI = sample estimate ± margin of error
  - Larger MOE → wider CI:
  - Means less precision in the estimate.
  - Smaller MOE → narrower CI:
  - Means more precise results.
  - MOE depends on:
  - Sample size (larger sample → smaller MOE)
  - Confidence level (higher confidence → larger MOE)
  - Standard deviation (more variability → larger MOE)

16. How is Bayes' Theorem used in statistics, and what is its significance
  - Updates Probabilities with New Data: Bayes' Theorem helps revise the probability of an event after getting new evidence.
  - Applies to Conditional Probability: It calculates the probability of an event A given B has happened, written as P(A|B).
  - Used in Hypothesis Testing: Helps in updating belief about a hypothesis when new sample data is observed.
  - Significance of Bayes' Theorem:
  - Foundation of Bayesian Statistics.
  - Incorporates prior knowledge with new data.
  - Improves decision-making in uncertain conditions.
  - Allows dynamic probability updating as new data comes in.

17. What is the Chi-square distribution, and when is it used
  - A Chi-square (χ²) distribution is a right-skewed probability distribution that arises when you sum the squares of independent standard normal variables.
  - It is only defined for non-negative values (x ≥ 0).
  - Depends on degrees of freedom (df) — more df → more symmetric the curve.
  - Chi-square Test for Goodness-of-Fit : To check if observed data fits a known or expected distribution.
  - Chi-square Test for Independence : To test if two categorical variables are independent in a contingency table.
  - Chi-square Test for Homogeneity : To compare the distribution of a variable across different groups.
  - In Confidence Intervals for Variance : Used to calculate confidence intervals and tests for population variance.

18.  What is the Chi-square goodness of fit test, and how is it applied
  - The Chi-square goodness-of-fit test checks whether a set of observed frequencies matches the expected frequencies based on a specific theoretical distribution.
  - Steps to Apply It:
  - State the hypotheses
  - Calculate the expected frequencies for each category.
  - Use the Chi-square formula
  - Determine the degrees of freedom (df)
  - Find the critical value or p-value from the Chi-square distribution table
  - Make a decision

19. What is the F-distribution, and when is it used in hypothesis testing
  - The F-distribution is a right-skewed probability distribution used mainly to compare variances. It is formed by the ratio of two chi-square distributions, each divided by their degrees of freedom.
  - ANOVA (Analysis of Variance): To test whether three or more population means are significantly different. Uses F-statistic to compare between-group variance to within-group variance.
  - Test of Equality of Variances: To check if two populations have the same variance.
  - Regression Analysis: To test the overall significance of a regression model.Compares explained vs. unexplained variance.

20. What is an ANOVA test, and what are its assumptions
  - ANOVA (Analysis of Variance) is a statistical method used to compare the means of three or more groups to determine if at least one group mean is significantly different from the others.
  - It uses the F-distribution to test the hypothesis.
  - It’s an extension of the t-test when comparing more than two groups.
  - Assumptions:
  - Data in each group must be collected independently
  - The data in each group should be approximately normally distributed.
  - All groups should have similar variances (use Levene’s Test or Bartlett’s Test to check).

21.  What are the different types of ANOVA tests
  -  One-Way ANOVA : Compare means of 3 or more groups based on one independent variable.
  - Two-Way ANOVA : Compare group means based on two independent variables, and also test for interaction effects.
  - Repeated Measures ANOVA : Used when the same subjects are measured multiple times (within-subject design).
  - Two-Way Repeated Measures ANOVA : Two factors are tested, and at least one is repeated (within-subjects).
  - MANOVA (Multivariate ANOVA) : Like ANOVA, but tests for multiple dependent variables simultaneously.

22. What is the F-test, and how does it relate to hypothesis testing
  - The F-test is a statistical test that compares two variances to determine if they are significantly different. It uses the F-distribution, which is the ratio of two chi-square distributions.
  - Test if two populations have equal variances.
  - Used in ANOVA to compare more than two group means.
  - Used in regression to test if the overall regression model is significant.

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

```python
def bernoulli_pdf(x, p):

    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    return p**x * (1 - p)**(1 - x)

# Example usage
print("P(X=1):", bernoulli_pdf(1, 0.7))
print("P(X=0):", bernoulli_pdf(0, 0.7))
```

9. Write a Python function to calculate Z-scores from a dataset and plot them

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_and_plot_z_scores(data):

    # Step 1: Convert data to NumPy array
    data = np.array(data)

    # Step 2: Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Step 3: Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Step 4: Plot original data vs Z-scores
    plt.figure(figsize=(12, 6))

    # Plot original data
    plt.subplot(1, 2, 1)
    sns.histplot(data, kde=True, color='skyblue')
    plt.title("Original Data Distribution")
    plt.xlabel("Data Values")
    plt.ylabel("Frequency")

    # Plot Z-scores
    plt.subplot(1, 2, 2)
    sns.histplot(z_scores, kde=True, color='salmon')
    plt.title("Z-Score Distribution")
    plt.xlabel("Z-scores")
    plt.ylabel("Frequency")

    plt.tight_layout()
    plt.show()

    return z_scores

# Example dataset
sample_data = [10, 12, 14, 11, 13, 15, 12, 14, 16, 18, 20]

# Calculate and plot Z-scores
z = calculate_and_plot_z_scores(sample_data)
print("Z-scores:", z)
```

10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def demonstrate_clt(sample_size=30, num_samples=1000, original_dist_lambda=1.0):

    # Step 1: Draw samples and compute means
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.exponential(scale=1/original_dist_lambda, size=sample_size)
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)

    # Step 2: Plot original distribution vs sample means
    plt.figure(figsize=(14, 6))

    # Original population distribution (exponential)
    plt.subplot(1, 2, 1)
    original_data = np.random.exponential(scale=1/original_dist_lambda, size=10000)
    sns.histplot(original_data, kde=True, color='orange')
    plt.title("Original Exponential Distribution")
    plt.xlabel("Value")
    plt.ylabel("Frequency")

    # Distribution of sample means
    plt.subplot(1, 2, 2)
    sns.histplot(sample_means, kde=True, color='green')
    plt.title(f"Distribution of Sample Means\n(Sample Size = {sample_size}, Samples = {num_samples})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")

    plt.tight_layout()
    plt.show()

demonstrate_clt(sample_size=50, num_samples=1000)
```

11. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def verify_clt_with_normal_distribution(sample_size=30, num_samples=1000, mu=50, sigma=10):

    # Step 1: Generate multiple samples and calculate their means
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.normal(loc=mu, scale=sigma, size=sample_size)
        sample_means.append(np.mean(sample))

    # Step 2: Plot original distribution
    plt.figure(figsize=(14, 6))

    # Original population distribution
    plt.subplot(1, 2, 1)
    original_data = np.random.normal(loc=mu, scale=sigma, size=10000)
    sns.histplot(original_data, kde=True, color='skyblue')
    plt.title("Original Normal Distribution")
    plt.xlabel("Value")
    plt.ylabel("Frequency")

    # Distribution of sample means
    plt.subplot(1, 2, 2)
    sns.histplot(sample_means, kde=True, color='seagreen')
    plt.title(f"Distribution of Sample Means\n(Sample Size = {sample_size}, Samples = {num_samples})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")

    plt.tight_layout()
    plt.show()

    # Print stats
    print(f"Original Mean (μ): {mu}, Sample Means Average: {np.mean(sample_means):.2f}")
    print(f"Original Std Dev (σ): {sigma}, Sample Means Std Dev: {np.std(sample_means):.2f}")

verify_clt_with_normal_distribution(sample_size=40, num_samples=1000)
```

distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def calculate_z_score(x, mu, sigma):

    return (x - mu) / sigma

def plot_z_score_on_standard_normal(z_score):


    # Generate x values for the standard normal curve
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, 0, 1)

    # Plot the standard normal curve
    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Standard Normal Distribution', color='blue')

    # Shade the area under the curve up to the Z-score
    plt.fill_between(x, 0, y, where=(x <= z_score), color='orange', alpha=0.5, label=f'Z ≤ {z_score:.2f}')

    # Draw a vertical line at the Z-score
    plt.axvline(z_score, color='red', linestyle='--', label=f'Z = {z_score:.2f}')
    plt.axvline(0, color='black', linestyle=':', label='Mean (Z = 0)')

    plt.title("Z-Score on Standard Normal Distribution")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
x = 85            # data point
mu = 70           # mean of the dataset
sigma = 10        # standard deviation

z = calculate_z_score(x, mu, sigma)
print(f"Z-score for x = {x} (mean = {mu}, std dev = {sigma}): {z:.2f}")

plot_z_score_on_standard_normal(z)
```

15. Implement hypothesis testing using Z-statistics for a sample dataset

```python
import numpy as np
import scipy.stats as stats

def z_test(sample_data, population_mean, population_std, alpha=0.05):


    # Sample statistics
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Z-score calculation
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Two-tailed p-value
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Critical value for two-tailed test
    z_critical = stats.norm.ppf(1 - alpha / 2)

    # Print results
    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-score = {z_score:.4f}")
    print(f"P-value = {p_value:.4f}")
    print(f"Z-critical = ±{z_critical:.4f}")

    # Decision
    if abs(z_score) > z_critical:
        print("Result: Reject the null hypothesis (H₀).")
    else:
        print("Result: Fail to reject the null hypothesis (H₀).")

    return z_score, p_value

# Sample data: weights of 30 chocolate bars
sample = [51, 49, 52, 50, 48, 49, 51, 50, 52, 53,
          50, 49, 51, 48, 47, 50, 52, 49, 51, 50,
          48, 49, 50, 52, 53, 51, 49, 48, 50, 50]

population_mean = 50      # Claimed average weight
population_std = 2        # Known population standard deviation

z_test(sample, population_mean, population_std, alpha=0.05)
```

16. Create a confidence interval for a dataset using Python and interpret the result

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):


    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard Error of the Mean

    # t-critical value (for unknown population std dev)
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Margin of Error
    margin = t_crit * std_err

    lower_bound = mean - margin
    upper_bound = mean + margin

    print(f"Sample Mean = {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    return lower_bound, upper_bound

# Example dataset: test scores of students
data = [82, 85, 88, 90, 87, 84, 91, 89, 86, 83]

confidence_interval(data, confidence=0.95)
```

17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

```python
import numpy as np
import scipy.stats as stats

def generate_and_ci_normal(n=50, mu=100, sigma=15, confidence=0.95):

    # Step 1: Generate data
    data = np.random.normal(loc=mu, scale=sigma, size=n)

    # Step 2: Calculate sample statistics
    sample_mean = np.mean(data)
    sample_std_err = stats.sem(data)  # Standard error of the mean

    # Step 3: Get t-critical value
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Step 4: Compute margin of error and confidence interval
    margin = t_crit * sample_std_err
    lower_bound = sample_mean - margin
    upper_bound = sample_mean + margin

    # Step 5: Print results
    print(f"Generated {n} samples from N({mu}, {sigma}²)")
    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"{int(confidence * 100)}% Confidence Interval for the Mean: ({lower_bound:.2f}, {upper_bound:.2f})")

    return sample_mean, lower_bound, upper_bound

generate_and_ci_normal(n=50, mu=100, sigma=15, confidence=0.95)
```

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results

```python
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, population_mean, population_std, alpha=0.05):

    # Step 1: Calculate sample statistics
    sample_mean = np.mean(sample)
    n = len(sample)

    # Step 2: Calculate Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Step 3: Calculate p-value for two-tailed test
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Step 4: Determine test result
    conclusion = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"

    # Step 5: Print results
    print(f"Sample Mean = {sample_mean:.4f}")
    print(f"Z-score = {z_score:.4f}")
    print(f"P-value = {p_value:.4f}")
    print(f"Conclusion: {conclusion} at α = {alpha}")

    return z_score, p_value, conclusion

# Sample data
sample = [101, 98, 100, 99, 102, 97, 103, 100, 99, 98]

# Population parameters
population_mean = 100
population_std = 2

one_sample_z_test(sample, population_mean, population_std)
```

4.  Perform a two-tailed Z-test using Python and visualize the decision region on a plot

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test_and_plot(sample, population_mean, population_std, alpha=0.05):

    # Step 1: Calculate sample statistics
    sample_mean = np.mean(sample)
    n = len(sample)
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Step 2: Z critical values
    z_critical = norm.ppf(1 - alpha / 2)

    # Step 3: Print results
    print(f"Sample Mean = {sample_mean:.4f}")
    print(f"Z-score = {z_score:.4f}")
    print(f"P-value = {p_value:.4f}")
    print(f"Z-critical values: ±{z_critical:.4f}")

    if abs(z_score) > z_critical:
        print("Result: Reject the null hypothesis")
    else:
        print("Result: Fail to reject the null hypothesis")

    # Step 4: Plot standard normal curve
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Standard Normal Distribution', color='blue')

    # Fill critical regions
    plt.fill_between(x, y, where=(x < -z_critical), color='red', alpha=0.3, label='Rejection Region')
    plt.fill_between(x, y, where=(x > z_critical), color='red', alpha=0.3)

    # Z-score line
    plt.axvline(z_score, color='black', linestyle='--', label=f'Z-score = {z_score:.2f}')

    # Labels and formatting
    plt.title("Two-Tailed Z-Test: Decision Regions")
    plt.xlabel("Z")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Sample data
sample_data = [101, 98, 100, 99, 102, 97, 103, 100, 99, 98]

# Known population parameters
population_mean = 100
population_std = 2

# Run the test and plot
two_tailed_z_test_and_plot(sample_data, population_mean, population_std, alpha=0.05)
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=50, alpha=0.05):

    # Step 1: Calculate standard error and critical value
    se = sigma / np.sqrt(n)
    z_alpha = norm.ppf(1 - alpha)  # one-tailed (right)
    critical_value = mu0 + z_alpha * se

    # Step 2: Define X-axis
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # PDFs
    y0 = norm.pdf(x, loc=mu0, scale=se)  # under H0
    y1 = norm.pdf(x, loc=mu1, scale=se)  # under H1

    # Step 3: Plot distributions
    plt.figure(figsize=(12, 6))
    plt.plot(x, y0, label='H₀: μ = {}'.format(mu0), color='blue')
    plt.plot(x, y1, label='H₁: μ = {}'.format(mu1), color='green')

    # Fill Type I Error Region (α): area in H₀'s tail beyond critical value
    plt.fill_between(x, y0, where=(x >= critical_value), color='red', alpha=0.3, label='Type I Error (α)')

    # Fill Type II Error Region (β): area in H₁ to the left of critical value
    plt.fill_between(x, y1, where=(x < critical_value), color='orange', alpha=0.4, label='Type II Error (β)')

    # Vertical line at critical value
    plt.axvline(critical_value, color='black', linestyle='--', label='Critical Value')

    # Labels and formatting
    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Step 4: Calculate beta (Type II error)
    z_beta = (critical_value - mu1) / se
    beta = norm.cdf(z_beta)
    power = 1 - beta

    # Print summary
    print(f"Critical Value: {critical_value:.2f}")
    print(f"Type I Error (α): {alpha:.2f}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the Test: {power:.4f}")

visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=50, alpha=0.05)
```

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):

    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)  # Standard Error of the Mean
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n - 1)  # T-critical value

    margin_of_error = t_crit * sem
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    # Output the result
    print(f"Sample Mean = {mean:.2f}")
    print(f"Confidence Level = {int(confidence * 100)}%")
    print(f"T-critical value = {t_crit:.4f}")
    print(f"Margin of Error = {margin_of_error:.4f}")
    print(f"Confidence Interval = ({lower_bound:.2f}, {upper_bound:.2f})")

    return lower_bound, upper_bound

# Simulated sample data
sample_data = [102, 100, 98, 101, 99, 100, 97, 103, 98, 100]

# Calculate 95% confidence interval
confidence_interval(sample_data, confidence=0.95)
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
def bayes_theorem(prior_H, likelihood_D_given_H, likelihood_D_given_not_H):

    # Prior for not having the disease
    prior_not_H = 1 - prior_H

    # Total probability of data (evidence)
    evidence = (likelihood_D_given_H * prior_H) + (likelihood_D_given_not_H * prior_not_H)

    # Posterior using Bayes’ Theorem
    posterior = (likelihood_D_given_H * prior_H) / evidence

    # Print intermediate results
    print(f"Prior (P(H))                = {prior_H:.4f}")
    print(f"Likelihood (P(D|H))         = {likelihood_D_given_H:.4f}")
    print(f"Likelihood (P(D|~H))        = {likelihood_D_given_not_H:.4f}")
    print(f"Evidence (P(D))             = {evidence:.4f}")
    print(f"Posterior (P(H|D))          = {posterior:.4f}")

    return posterior

# Inputs for the disease test problem
prior = 0.01              # 1% prevalence
likelihood_given_disease = 0.95
likelihood_given_no_disease = 0.05

bayes_theorem(prior, likelihood_given_disease, likelihood_given_no_disease)
```

12.  Perform a Chi-square test for independence between two categorical variables in Python

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

def chi_square_test_of_independence():


    # Step 1: Create a contingency table (example: Gender vs Preference)
    data = {
        'Likes Product': [30, 10],     # e.g., Male, Female
        'Dislikes Product': [20, 40]
    }
    index = ['Male', 'Female']
    table = pd.DataFrame(data, index=index)

    print("Contingency Table:")
    print(table, "\n")

    # Step 2: Perform the Chi-square test
    chi2, p, dof, expected = chi2_contingency(table)

    # Step 3: Output the results
    print(f"Chi-square Statistic = {chi2:.4f}")
    print(f"P-value = {p:.4f}")
    print(f"Degrees of Freedom = {dof}")
    print("Expected Frequencies:")
    print(pd.DataFrame(expected, index=index, columns=table.columns))

    # Step 4: Conclusion
    alpha = 0.05
    if p < alpha:
        print("\nConclusion: Reject the null hypothesis → Variables are associated.")
    else:
        print("\nConclusion: Fail to reject the null hypothesis → Variables are independent.")


chi_square_test_of_independence()
```

13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data

```python
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):

    # Convert to DataFrame if not already
    if not isinstance(observed, pd.DataFrame):
        observed = pd.DataFrame(observed)

    # Step 1: Compute row totals, column totals, and grand total
    row_totals = observed.sum(axis=1)
    col_totals = observed.sum(axis=0)
    grand_total = observed.values.sum()

    # Step 2: Compute expected frequencies
    expected = pd.DataFrame(index=observed.index, columns=observed.columns)

    for i in observed.index:
        for j in observed.columns:
            expected.loc[i, j] = (row_totals[i] * col_totals[j]) / grand_total

    expected = expected.astype(float)

    # Output
    print("Observed Frequencies:\n", observed)
    print("\nExpected Frequencies:\n", expected)

    return expected

# Sample observed data: Gender vs Preference
observed_data = pd.DataFrame({
    'Likes Product': [30, 10],
    'Dislikes Product': [20, 40]
}, index=['Male', 'Female'])

calculate_expected_frequencies(observed_data)
```

14.  Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution

```python
import numpy as np
from scipy.stats import chisquare

def goodness_of_fit_test(observed, expected):


    # Perform the test using scipy
    chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

    # Print results
    print("Observed Frequencies:", observed)
    print("Expected Frequencies:", expected)
    print(f"Chi-square Statistic = {chi2_stat:.4f}")
    print(f"P-value = {p_value:.4f}")

    # Interpret result
    alpha = 0.05
    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis → The data does not fit the expected distribution.")
    else:
        print("Conclusion: Fail to reject the null hypothesis → The data fits the expected distribution.")

    return chi2_stat, p_value

# Example: Observed number of people choosing 4 different flavors
observed = [50, 30, 15, 5]

# Expected proportions: 40%, 30%, 20%, 10% → expected counts for 100 samples
expected = [40, 30, 20, 10]

goodness_of_fit_test(observed, expected)
```
